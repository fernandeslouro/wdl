```python
import pandas as pd
```

```python
population_dataframe = pd.DataFrame(columns=pd.read_csv('lisboa.csv', encoding='Latin-1', sep=';').columns)
for region in ['centro', 'lisboa', 'norte', 'alentejo', 'algarve']:
    population_dataframe = population_dataframe.append(pd.read_csv(f'{region}.csv', encoding='Latin-1', sep=';'), ignore_index=True)
    
population_dataframe = population_dataframe[population_dataframe.nivel==6]
population_dataframe['Zona Geográfica'] = population_dataframe['Zona Geográfica'].str.lstrip()
```

```python
population_dataframe.to_csv('parishes_population.csv', index=False)
```
