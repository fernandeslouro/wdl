```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tf = pd.read_csv('traffic_flow_2021.csv', encoding = "ISO-8859-1")

tf = tf.drop(columns=['dataprovider','source', 'entity_type', 'name', 'dateobservedto', 'laneid', 'time_from', 'lanedirection'])

tf.entity_id=tf.entity_id.str[52:]
```

```python
tf
```

```python
d = tf[tf.date_from == '2021-03-15']
```

```python
tp = d[d.entity_id == '16Z2'].sort_values('dateobservedfrom')
```

```python
for ent in tf.entity_id.unique()[6:10]:
    pt = d[d.entity_id == ent].sort_values('dateobservedfrom')
    #pt = sg.groupby('date_from').agg(['max', 'min', 'mean']).sort_values('date_from')
    plt.figure(figsize=(10, 2))
    plt.title(ent)
    plt.axvline('2021-03-15 05:00:00')
    plt.axvline('2021-03-15 22:00:00')
    plt.plot(pt.dateobservedfrom, pt.intensity)
    plt.plot(pt.dateobservedfrom, pt.intensity)
```
